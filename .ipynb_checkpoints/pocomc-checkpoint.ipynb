```python
# PoCoMC applied to planet-only vs planet+moon light curve models

import numpy as np
import matplotlib.pyplot as plt
import pocomc as pc
from scipy.stats import uniform
import corner
import batman
```

```python
# Load your data: time, observed flux, and flux uncertainties
data = np.loadtxt('/pscratch/sd/e/emmayu/kipping_exomoons/KIC-3239945/photometry/planet1/AVG_SAP.dat')
time_array = data[:,0]
observed_flux = data[:,1]
flux_err = data[:,2]

```

```python
def generate_mandel_agol_model(P, tau, p, b, rho_star, q1, q2, time_array):
    """
    Parameters:
    - P: orbital period [days]
    - tau: transit time (mid-transit time) [same units as time_array]
    - p: Rp/Rs (planet radius / star radius)
    - b: impact parameter (unitless)
    - rho_star: stellar density [g/cm^3 or solar units, converted below]
    - q1, q2: quadratic limb darkening coefficients in Kipping's parameterization
    - time_array: 1D numpy array of time stamps [days]
    
    Returns:
    - flux: model flux at each time point
    """

    # Constants
    G = 6.67430e-11       # gravitational constant (m^3 kg^-1 s^-2)
    Msun = 1.98847e30     # solar mass (kg)
    Rsun = 6.957e8        # solar radius (m)
    day = 86400.0         # seconds in a day
    
    # Convert stellar density rho_star to SI (kg/m^3) if needed
    # Here assume rho_star is in g/cm^3:
    rho_star_SI = rho_star * 1000  # 1 g/cm3 = 1000 kg/m3
    
    # Calculate semi-major axis in stellar radii (a/Rs) using Kepler's Third Law:
    # a^3 = (G * M_star * P^2) / (4 * pi^2)
    # M_star can be estimated from rho_star and Rs:
    # rho_star = M_star / (4/3 pi Rs^3)
    # => M_star = rho_star * (4/3) * pi * Rs^3
    
    M_star = rho_star_SI * (4/3) * np.pi * Rsun**3  # kg
    a_meters = ((G * M_star * (P * day)**2) / (4 * np.pi**2))**(1/3)
    a_rs = a_meters / Rsun  # semi-major axis in stellar radii
    
    # Calculate inclination from impact parameter: b = (a/Rs) * cos(i)
    incl_rad = np.arccos(b / a_rs)
    incl_deg = np.degrees(incl_rad)
    
    # Convert limb darkening parameters q1, q2 (Kipping 2013) to u1, u2
    u1 = 2 * np.sqrt(q1) * q2
    u2 = np.sqrt(q1) * (1 - 2 * q2)
    
    # Setup batman parameters
    params = batman.TransitParams()
    params.t0 = tau         # time of inferior conjunction (mid-transit)
    params.per = P          # orbital period in days
    params.rp = p           # planet radius (in stellar radii)
    params.a = a_rs         # semi-major axis (in stellar radii)
    params.inc = incl_deg   # orbital inclination in degrees
    params.ecc = 0.0        # eccentricity (assume circular)
    params.w = 90.0         # longitude of periastron (degrees)
    params.u = [u1, u2]     # quadratic limb darkening coefficients
    params.limb_dark = "quadratic"
    
    # Create model
    m = batman.TransitModel(params, time_array)
    flux = m.light_curve(params)
    
    return flux

```

```python
prior_bounds = {
    'P':        (1000, 1100),      # orbital period
    'tau':      (55250, 55260),    # mid-transit time
    'p':        (0.01, 0.2),       # Rp/R*
    'b':        (0.0, 1.0),        # impact parameter
    'rho_star': (0.1, 5.0),        # stellar density
    'q1':       (0.0, 1.0),        # limb darkening param 1
    'q2':       (0.0, 1.0),        # limb darkening param 2
}
```

```python
param_names = list(prior_bounds.keys())
n_dim = len(param_names)

# === Define prior ===
prior_distributions = [uniform(loc=low, scale=high - low) for low, high in prior_bounds.values()]
prior = pc.Prior(prior_distributions)
```

```python
# === Define log-likelihood ===
def log_likelihood_planet(x):
    """
    x: (n_samples, n_params) array of parameter vectors
    Returns: (n_samples,) array of log-likelihoods
    """
    P, tau, p, b, rho, q1, q2 = x.T  # unpack columns

    # Evaluate model for each sample (slow loop unless model is vectorized)
    model_fluxes = np.array([
        generate_mandel_agol_model(Pi, taui, pi, bi, rhoi, q1i, q2i, time_array)
        for Pi, taui, pi, bi, rhoi, q1i, q2i in zip(P, tau, p, b, rho, q1, q2)
    ])

    # Chi-squared for each model
    chi2 = np.sum(((model_fluxes - observed_flux) / flux_err) ** 2, axis=1)

    return -0.5 * chi2
```

```python
# === Initialize pocoMC Sampler ===
sampler = pc.Sampler(
    prior=prior,
    likelihood=log_likelihood_planet,
    vectorize=True,
    random_state=0
)
```

```python
# === Run the sampler ===

sampler.run()

```

```python
# === Get results ===
samples, weights, logl, logp = sampler.posterior()
logZ, logZ_err = sampler.evidence()

print(f"[Planet-only] logZ = {logZ:.3f} ± {logZ_err:.3f}")
```

```python
# Get outputs
samples, weights, logl, logp = sampler.posterior()
logZ, logZ_err = sampler.evidence()

# Save everything in a single file
np.savez_compressed("planet_only_pocomc_results.npz",
    samples=samples,
    weights=weights,
    logl=logl,
    logp=logp,
    logZ=logZ,
    logZ_err=logZ_err,
    param_names=np.array(["P", "tau", "p", "b", "rho_star", "q1", "q2"])
)
```

```python
# === Plot corner ===
corner.corner(samples, weights=weights, labels=param_names)
plt.show()
```

```python

# Load results
data = np.load("planet_only_pocomc_results.npz")
samples = data["samples"]
weights = data["weights"]

# Normalize weights if not already
weights /= np.sum(weights)

# Option 1: Weighted median sample (smoother)
median_params = np.average(samples, axis=0, weights=weights)

# Option 2: Maximum a posteriori (highest weight)
map_params = samples[np.argmax(weights)]

# Choose one
best_fit_params = median_params  # or use map_params
```

```python
# Unpack
P, tau, p, b, rho_star, q1, q2 = best_fit_params

# Generate model
model_flux = generate_mandel_agol_model(P, tau, p, b, rho_star, q1, q2, time_array)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.errorbar(time_array, observed_flux, yerr=flux_err, fmt='.', color='black', alpha=0.5, label="Observed")
plt.plot(time_array, model_flux, color='blue', lw=2, label="PocoMC Posterior Model")
plt.xlabel("Time")
plt.ylabel("Flux")
plt.title("Transit Light Curve: First Epoch")
plt.xlim(55252.5, 55254)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load posterior
data = np.load("planet_only_pocomc_results.npz")
samples = data["samples"]
weights = data["weights"]
weights /= np.sum(weights)  # normalize weights

# Number of light curves to draw
n_draws = 100

# Time array and data should already exist
# time_array, observed_flux, flux_err
# generate_mandel_agol_model(...) should be defined

# === Plot ===
plt.figure(figsize=(10, 5))

# Plot model light curves from posterior draws
for i in range(n_draws):
    # Draw sample index according to posterior weights
    idx = np.random.choice(len(samples), p=weights)
    sample = samples[idx]
    
    # Unpack parameters
    P, tau, p, b, rho_star, q1, q2 = sample

    # Generate model flux
    model_flux = generate_mandel_agol_model(P, tau, p, b, rho_star, q1, q2, time_array)

    # Plot it with low opacity
    plt.plot(time_array, model_flux, color="blue", alpha=0.05)

# Plot observed data
plt.errorbar(time_array, observed_flux, yerr=flux_err, fmt=".", color="black", alpha=0.4, label="Observed", lw=0.5)

plt.xlabel("Time")
plt.ylabel("Flux")
plt.title(f"{n_draws} Posterior Light Curves")
plt.legend()
plt.xlim(55252.5, 55254)
plt.tight_layout()
plt.grid(True)
plt.show()
```
