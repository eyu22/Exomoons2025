```python
import matplotlib.pyplot as plt
import numpy as np

# Load data from file (two columns: z, flux)
z, flux = np.loadtxt("/global/homes/e/emmayu/Exomoons2025/LUNA/lightcurve.txt", unpack=True)

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(z, flux, marker='o', linestyle='-', color='blue')
plt.xlabel("Impact parameter (z)")
plt.ylabel("Flux")
plt.title("Transit Light Curve: Flux vs Impact Parameter")
plt.grid(True)
plt.tight_layout()
plt.show()
```
